```python
import os
import time
import requests
from tqdm.notebook import tqdm

url = "https://huggingface.co/datasets/McAuley-Lab/Amazon-Reviews-2023/resolve/main/raw/review_categories/Electronics.jsonl"

BASE_DIR = r"C:\Users\dedee\Downloads\NLP"
os.makedirs(BASE_DIR, exist_ok=True)
output_path = os.path.join(BASE_DIR, "Electronics.jsonl")

CHUNK_SIZE = 512 * 1024      # 512 KB (smaller = fewer timeouts)
TIMEOUT = 15                # seconds
MAX_RETRIES = 5

downloaded = os.path.getsize(output_path) if os.path.exists(output_path) else 0
headers = {"Range": f"bytes={downloaded}-"} if downloaded else {}

print(f"Resuming from {downloaded / (1024**2):.2f} MB")
#retry logic

for attempt in range(1, MAX_RETRIES + 1):
    try:
        with requests.get(
            url,
            headers=headers,
            stream=True,
            timeout=TIMEOUT
        ) as response:
            response.raise_for_status()

            total_size = int(response.headers.get("Content-Length", 0)) + downloaded

            with open(output_path, "ab") as f, tqdm(                                                                #append binary
                total=total_size,
                initial=downloaded,
                unit="B",
                unit_scale=True,
                desc="Downloading Electronics.jsonl"
            ) as pbar:
                for chunk in response.iter_content(chunk_size=CHUNK_SIZE):
                    if chunk:
                        f.write(chunk)
                        pbar.update(len(chunk))

        print(" Download completed successfully")
        break

    except requests.exceptions.RequestException as e:
        print(f" Timeout / network error (attempt {attempt}/{MAX_RETRIES})")
        time.sleep(5)

else:
    print(" Download failed after maximum retries")
```

# N-GRAM ANALYSIS (Unigrams, Bigrams, Trigrams)

```python
def get_top_ngrams(corpus, ngram_range=(1,1), top_n=15):                                                                    #convert into bag of words matrix
    vec = CountVectorizer(
        stop_words="english",
        ngram_range=ngram_range,
        max_features=5000
    )
    X = vec.fit_transform(corpus)
    sum_words = X.sum(axis=0)

    words_freq = [
        (word, sum_words[0, idx])
        for word, idx in vec.vocabulary_.items()
    ]

    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    return words_freq[:top_n]


```

# TEXT CLEANING FUNCTION

```python
def clean_text(text):
    if not isinstance(text, str):
        return ""

    # Normalize unicode
    text = unicodedata.normalize("NFKD", text)

    # Remove HTML tags
    text = re.sub(r"<.*?>", "", text)

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emojis
    text = emoji.replace_emoji(text, replace="")

    # Remove special characters
    text = re.sub(r"[^a-zA-Z0-9\s.,!?]", "", text)

    # Lowercase
    text = text.lower()

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
def remove_html(text):
    if pd.isna(text):
        return ""
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text(separator=" ")
```

```python
def remove_emojis(text):
    return emoji.replace_emoji(text, replace="")
```

```python
def normalize_whitespace(text):
    text = re.sub(r"\s+", " ", text)   # multiple spaces → single space
    return text.strip()
```

```python
import pandas as pd

train_df = pd.read_csv("electronics_5k_train.csv")
val_df   = pd.read_csv("electronics_5k_val.csv")

train_df = train_df[['text_clean', 'title_clean']]
val_df   = val_df[['text_clean', 'title_clean']]

print("Train:", train_df.shape)
print("Val:", val_df.shape)
```

```python
import torch
from transformers import BartTokenizer, BartForConditionalGeneration

MODEL_NAME = "facebook/bart-large-cnn"

tokenizer = BartTokenizer.from_pretrained(MODEL_NAME)
model = BartForConditionalGeneration.from_pretrained(MODEL_NAME)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("BART loaded")
```

Tokenization

```python
MAX_INPUT  = 512
MAX_TARGET = 64

def tokenize_bart(batch):
    # Encode the source text
    inputs = tokenizer(
        batch["text_clean"],
        truncation=True,
        padding="max_length",
        max_length=MAX_INPUT
    )

    # Encode the target text (summary) directly
    labels = tokenizer(
        batch["title_clean"],
        truncation=True,
        padding="max_length",
        max_length=MAX_TARGET
    )

    # Replace pad token id with -100 for loss computation
    labels["input_ids"] = [
        [(l if l != tokenizer.pad_token_id else -100) for l in label]
        for label in labels["input_ids"]
    ]

    # Assign labels to inputs
    inputs["labels"] = labels["input_ids"]

    return inputs
```

```python
from datasets import Dataset

train_ds = Dataset.from_pandas(train_df)
val_ds   = Dataset.from_pandas(val_df)

train_ds = train_ds.map(tokenize_bart, batched=True, remove_columns=train_ds.column_names)
val_ds   = val_ds.map(tokenize_bart, batched=True, remove_columns=val_ds.column_names)
```

```python
from transformers import TrainingArguments, Trainer
from transformers import DataCollatorForSeq2Seq
import torch

training_args = TrainingArguments(
    output_dir="./bart_finetuned_clean",
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    learning_rate=5e-5,
    num_train_epochs=2,
    logging_steps=50,
    save_total_limit=2,
    fp16=torch.cuda.is_available(),
    report_to="none"
)

data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    model=model
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    data_collator=data_collator
)
```

```python
trainer.train()
model.eval()
```

Generate Summaries

```python
from tqdm import tqdm

MAX_INPUT = 512
MAX_TARGET = 64

def generate_summary(text):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding="max_length",
        max_length=MAX_INPUT
    ).to(device)

    summary_ids = model.generate(
        **inputs,
        max_length=MAX_TARGET,
        min_length=20,
        num_beams=4,
        length_penalty=2.0,
        early_stopping=True
    )

    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

val_df["bart_summary"] = [
    generate_summary(t) for t in tqdm(val_df["text_clean"], desc="Generating BART summaries")
]
```

```python
import torch
import pandas as pd
import numpy as np
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
train_df = pd.read_csv("electronics_5k_train.csv")
val_df   = pd.read_csv("electronics_5k_val.csv")

train_df = train_df[["text_clean", "title_clean"]]
val_df   = val_df[["text_clean", "title_clean"]]

train_ds = Dataset.from_pandas(train_df)
val_ds   = Dataset.from_pandas(val_df)
```

```python
MODEL_NAME = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME).to(device)
```

```python
MAX_INPUT = 512
MAX_TARGET = 64

def tokenize_t5(batch):
    # Prefix input with "summarize: "
    inputs = ["summarize: " + t for t in batch["text_clean"]]

    # Tokenize inputs
    model_inputs = tokenizer(
        inputs,
        max_length=MAX_INPUT,
        truncation=True,
        padding="max_length"
    )

    # Tokenize targets (summaries) directly
    labels = tokenizer(
        batch["title_clean"],
        max_length=MAX_TARGET,
        truncation=True,
        padding="max_length"
    )

    # Replace pad token ids by -100 for PyTorch loss
    labels["input_ids"] = [
        [(l if l != tokenizer.pad_token_id else -100) for l in label]
        for label in labels["input_ids"]
    ]

    # Assign labels to model inputs
    model_inputs["labels"] = labels["input_ids"]

    return model_inputs
```

```python
train_tokenized = train_ds.map(tokenize_t5, batched=True, remove_columns=train_ds.column_names)
val_tokenized   = val_ds.map(tokenize_t5, batched=True, remove_columns=val_ds.column_names)
```

```python
from transformers import TrainingArguments, Trainer, DataCollatorForSeq2Seq

training_args = TrainingArguments(
    output_dir="./t5_results_clean",
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=2,
    learning_rate=5e-5,
    logging_steps=50,
    save_total_limit=2,
    fp16=torch.cuda.is_available(),
    report_to="none"
)

data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    model=model
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_tokenized,
    eval_dataset=val_tokenized,
    data_collator=data_collator
)
```

```python
trainer.train()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

MODEL_DIR = "./bart_qlora_finetuned_colab"

tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_DIR)
model.to(device)
model.eval()
```

```python
def generate_summary(
    text,
    max_length=64,
    decoding_strategy="greedy",
    num_beams=4,
    temperature=1.0,
    top_k=50,
    top_p=0.95
):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=512
    ).to(device)

    if decoding_strategy == "greedy":
        output_ids = model.generate(
            **inputs,
            max_length=max_length
        )

    elif decoding_strategy == "beam":
        output_ids = model.generate(
            **inputs,
            max_length=max_length,
            num_beams=num_beams,
            early_stopping=True
        )

    elif decoding_strategy == "temperature":
        output_ids = model.generate(
            **inputs,
            max_length=max_length,
            do_sample=True,
            temperature=temperature
        )

    elif decoding_strategy == "top_k":
        output_ids = model.generate(
            **inputs,
            max_length=max_length,
            do_sample=True,
            top_k=top_k
        )

    elif decoding_strategy == "top_p":
        output_ids = model.generate(
            **inputs,
            max_length=max_length,
            do_sample=True,
            top_p=top_p
        )

    else:
        raise ValueError("Invalid decoding strategy")

    return tokenizer.decode(output_ids[0], skip_special_tokens=True)
```

```python
from datasets import load_dataset
from transformers import BartTokenizer

DATA_PATH = r"C:\Users\dedee\Downloads\NLP\electronics_5k_bart_summaries_clean(2).csv"

dataset = load_dataset("csv", data_files={"train": DATA_PATH})

tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")

def tokenize(batch):
    return tokenizer(
        batch["bart_summary"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

dataset = dataset.map(tokenize, batched=True)

# Convert ratings 1–5 → 0–4
def prepare_labels(batch):
    batch["labels"] = batch["rating"] - 1
    return batch

dataset = dataset.map(prepare_labels)

dataset.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)
```

```python
import torch
import torch.nn as nn
from transformers import BartModel

class BartRatingPredictor(nn.Module):
    def __init__(self, model_name="facebook/bart-large-cnn", num_labels=5):
        super().__init__()
        self.bart = BartModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(0.1)
        self.classifier = nn.Linear(self.bart.config.d_model, num_labels)
        self.loss_fn = nn.CrossEntropyLoss()

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.bart(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        hidden = outputs.last_hidden_state
        mask = attention_mask.unsqueeze(-1).float()

        pooled = (hidden * mask).sum(dim=1) / mask.sum(dim=1)
        pooled = self.dropout(pooled)

        logits = self.classifier(pooled)

        if labels is not None:
            loss = self.loss_fn(logits, labels)
            return {"loss": loss, "logits": logits}

        return {"logits": logits}
```

```python
import torch
import torch.nn as nn
from transformers import BartModel

class BartRatingPredictor(nn.Module):
    def __init__(self, model_name="facebook/bart-large-cnn", num_labels=5):
        super().__init__()
        
        self.bart = BartModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(0.1)
        self.classifier = nn.Linear(self.bart.config.d_model, num_labels)

    def forward(self, input_ids, attention_mask):
        outputs = self.bart(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        # Mean pooling
        hidden_states = outputs.last_hidden_state
        pooled = hidden_states.mean(dim=1)

        pooled = self.dropout(pooled)
        logits = self.classifier(pooled)

        return logits
```

```python
import torch
import torch.nn as nn
from transformers import BartModel
```

```python
class BartRatingPredictor(nn.Module):
    def __init__(self, model_name="facebook/bart-base", num_labels=5):
        super().__init__()

        self.encoder = BartModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(0.1)
        self.classifier = nn.Linear(
            self.encoder.config.d_model, num_labels
        )

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        # Mean pooling over tokens
        hidden_states = outputs.last_hidden_state
        pooled = hidden_states.mean(dim=1)

        pooled = self.dropout(pooled)
        logits = self.classifier(pooled)

        loss = None
        if labels is not None:
            loss_fn = nn.CrossEntropyLoss()
            loss = loss_fn(logits, labels)

        return {"loss": loss, "logits": logits}
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = BartRatingPredictor()
model.to(device)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, mean_absolute_error
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)

    acc = accuracy_score(labels, preds)
    mae = mean_absolute_error(labels, preds)

    return {
        "accuracy": acc,
        "mae": mae
    }
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np

from transformers import BertTokenizer, BertModel, Trainer, TrainingArguments
from datasets import Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv(
    "C:\\Users\\dedee\\Downloads\\NLP\\electronics_5k_bart_summaries_clean(2).csv"
)

df = df[["bart_summary", "rating"]]
df = df.dropna()

train_df, val_df = train_test_split(
    df, test_size=0.2, random_state=42
)
```

```python
MODEL_NAME = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)

MAX_LEN = 256

def tokenize(batch):
    enc = tokenizer(
        batch["bart_summary"],
        padding="max_length",
        truncation=True,
        max_length=MAX_LEN
    )
    enc["labels"] = batch["rating"]
    return enc
```

```python
train_ds = Dataset.from_pandas(train_df)
val_ds   = Dataset.from_pandas(val_df)

train_ds = train_ds.map(tokenize, batched=True)
val_ds   = val_ds.map(tokenize, batched=True)

train_ds.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)
val_ds.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)
```

```python
class BertRatingRegressor(nn.Module):
    def __init__(self):
        super().__init__()                      #pytorch nn base class
        self.bert = BertModel.from_pretrained("bert-base-uncased")
        self.dropout = nn.Dropout(0.1)
        self.regressor = nn.Linear(768, 1)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        pooled_output = outputs.pooler_output
        pooled_output = self.dropout(pooled_output)

        preds = self.regressor(pooled_output).squeeze(-1)

        loss = None
        if labels is not None:
            loss = nn.MSELoss()(preds, labels.float())

        return {"loss": loss, "logits": preds}
```

```python
model = BertRatingRegressor()
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./bert_rating_model",
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=2e-5,
    num_train_epochs=5,
    eval_strategy="epoch",     
    save_strategy="epoch",
    logging_steps=100,
    report_to="none",
)
```

```python
def compute_metrics(eval_pred):
    preds, labels = eval_pred
    mae = mean_absolute_error(labels, preds)
    rmse = np.sqrt(mean_squared_error(labels, preds))
    return {
        "MAE": mae,
        "RMSE": rmse
    }
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
trainer.save_model("./bert_rating_model")
tokenizer.save_pretrained("./bert_rating_model")
```

```python
def predict_rating(text):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding="max_length",
        max_length=MAX_LEN
    )

    with torch.no_grad():
        output = model(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"]
        )

    return float(output["logits"].item())
```

```python
df["predicted_rating"] = df["bart_summary"].apply(predict_rating)
df.to_csv("bert_rating_predictions.csv", index=False)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(df["rating"], df["predicted_rating"])
rmse = np.sqrt(mean_squared_error(df["rating"], df["predicted_rating"]))

print("MAE :", mae)
print("RMSE:", rmse)
```
